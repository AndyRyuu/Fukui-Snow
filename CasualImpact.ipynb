```python
import pandas as pd
from causalimpact import CausalImpact

# 创建时间序列数据
data = pd.DataFrame({'Sales': [10, 12, 8, 15, 20, 25, 18, 13, 17, 22],
                     'Date': pd.date_range(start='2022-01-01', periods=10)})

data = data.set_index('Date')
data
```

```python
# 创建活动时间段（假设第5天后开始了营销活动）
pre_period = ['2022-01-01', '2022-01-04']
post_period = ['2022-01-05', '2022-01-10']

# 创建CausalImpact对象并进行分析
ci = CausalImpact(data, pre_period, post_period)
ci.plot(figsize=(22, 20))
```
